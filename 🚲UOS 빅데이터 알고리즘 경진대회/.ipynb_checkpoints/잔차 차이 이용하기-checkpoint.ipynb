```python
def preprocess_data(data):
    data.set_index('일시', inplace=True)
    data.index = pd.to_datetime(data.index, format='%Y%m%d')
    data['week'] = data.index.isocalendar().week
    data['day_name'] = data.index.weekday
    data['weekend'] = data['day_name'].apply(lambda x: 0 if x < 5 else 1)
    data['year'] = data.index.year
    return data

data = preprocess_data(data)
sample_submission = preprocess_data(sample_submission)
```

```python
def smooth(y, box_pts):
    s = pd.Series(y)
    y_smooth = s.rolling(window=box_pts, min_periods=1, center=True).mean()
    return y_smooth.values
```

```python
from pytimekr import pytimekr
def add_holidays(data, start_year, end_year):
    holidays = []
    for year in range(start_year, end_year + 1):
        for holiday in pytimekr.holidays(year=year):
            if pytimekr.red_days(holiday) is not None:
                for day in pytimekr.red_days(holiday):
                    holidays.append(day.strftime('%Y-%m-%d'))
            else:
                holidays.append(holiday.strftime('%Y-%m-%d'))
    data.loc[holidays, 'weekend'] = 1
    return data
data = add_holidays(data, 2018, 2021)
```

```python
def plot_weekend_trends(data, gu):
    data_week_groupby = data.groupby(['year', 'week', 'weekend']).mean().reset_index()
    data_years = data_week_groupby['year'].unique()
    num_years = len(data_years)

    fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

    for i, year in enumerate(data_years):
        row = i // 2
        col = i % 2
        data_year = data_week_groupby[data_week_groupby['year'] == year].iloc[:, [1, 2, 3, 4, 5, 6]]
        data_weekend_1 = data_year[data_year['weekend'] == 1]
        data_weekend_0 = data_year[data_year['weekend'] == 0]

        axs[row, col].plot(data_weekend_1['week'], data_weekend_1[gu])
        axs[row, col].plot(data_weekend_0['week'], data_weekend_0[gu])
        axs[row, col].legend(['Weekend', 'Weekday'])
        axs[row, col].set_title(f'Trend for {gu} in {year}')
        axs[row, col].set_ylabel('Counts')

    plt.xlabel('Week')
    plt.tight_layout()
    plt.show()

plot_weekend_trends(data, '동대문구')
plot_weekend_trends(data, '성동구')
plot_weekend_trends(data, '광진구')
plot_weekend_trends(data, '중랑구')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_2018 = MinMaxScaler();scaler_2019 = MinMaxScaler()
scaler_2020 = MinMaxScaler();scaler_2021 = MinMaxScaler()
```

```python
year_df_list=[df_2018,df_2019,df_2020,df_2021]
scaler_list=[scaler_2018,scaler_2019,scaler_2020,scaler_2021]
for i in range(len(year_df_list)):
    year_df_list[i].iloc[:, 2:] = scaler_list[i].fit_transform(year_df_list[i].iloc[:, 2:])
```

```python
def it_weekmean(scaler, week_mean):
    return pd.DataFrame(
        scaler.inverse_transform(week_mean.iloc[:, 2:]),
        columns=week_mean.columns[2:]
    ).assign(week=week_mean['week'], weekend=week_mean['weekend'])

df_2018_weekmean = it_weekmean(scaler_2018, week_mean)
df_2019_weekmean = it_weekmean(scaler_2019, week_mean)
df_2020_weekmean = it_weekmean(scaler_2020, week_mean)
df_2021_weekmean = it_weekmean(scaler_2021, week_mean)
```

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 20))
colors = ['maroon', 'dodgerblue', 'violet', 'lime']

for i, ax in enumerate(axs.flat):
    for year, df_weekmean, color in zip(range(2018, 2022), 
                                        [df_2018_weekmean, df_2019_weekmean, df_2020_weekmean, df_2021_weekmean],
                                        colors):
        ax.plot(df_weekmean.iloc[:,i], label=f'{year}_{df_weekmean.columns[i]}', color=color)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
df_2022_weekmean=df_2021_weekmean.copy()
```

```python
for i in range(4):
    diff_mean = pd.concat([df_2018_weekmean.iloc[:,i], df_2019_weekmean.iloc[:,i], df_2020_weekmean.iloc[:,i], df_2021_weekmean.iloc[:,i]], axis=1).diff(axis=1).mean(axis=1)
    df_2022_weekmean.iloc[:,i] = df_2021_weekmean.iloc[:,i] + diff_mean * 3
```
