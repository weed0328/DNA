# 파생변수& 전처리

```python
train.drop(['FLAG_MOBIL'], axis=1, inplace=True)
test.drop(['FLAG_MOBIL'], axis=1, inplace=True)
# FLAG_MOBAIL 변수 삭제
```

```python
train['DAYS_EMPLOYED'] = train['DAYS_EMPLOYED'].map(lambda x: 0 if x > 0 else x)
test['DAYS_EMPLOYED'] = test['DAYS_EMPLOYED'].map(lambda x: 0 if x > 0 else x)
#임용기간 변수 추가
```

```python
train.fillna('NaN', inplace=True)
test.fillna('NaN', inplace=True)
```

```python
# 나이 만들기
train['new_age'] = round(abs(train['DAYS_BIRTH'])/365.5,0).astype(np.int32)
test['new_age'] = round(abs(test['DAYS_BIRTH'])/365.5,0).astype(np.int32)
```

```python
train['근속연수'] = train['DAYS_EMPLOYED'] // 365 # 근속연수
train['근속월수'] = train['DAYS_EMPLOYED'] // 30 # 근속월수
train['임용 월'] = np.floor(train['DAYS_EMPLOYED'] / 30) - ((np.floor(train['DAYS_EMPLOYED'] / 30) / 12).astype(int) * 12) # 고용된 달 
train['임용 주'] = np.floor(train['DAYS_EMPLOYED'] / 7) - ((np.floor(train['DAYS_EMPLOYED'] / 7) / 4).astype(int) * 4) # 고용된 주
train["고용전 날 수"] = train["DAYS_BIRTH"] - train["DAYS_EMPLOYED"]

test['근속연수'] = test['DAYS_EMPLOYED'] // 365 # 근속연수
test['근속월수'] = test['DAYS_EMPLOYED'] // 30 # 근속월수
test['임용 월'] = np.floor(test['DAYS_EMPLOYED'] / 30) - ((np.floor(test['DAYS_EMPLOYED'] / 30) / 12).astype(int) * 12) # 고용된 달 
test['임용 주'] = np.floor(test['DAYS_EMPLOYED'] / 7) - ((np.floor(test['DAYS_EMPLOYED'] / 7) / 4).astype(int) * 4) # 고용된 주
test["고용전 날 수"] = test["DAYS_BIRTH"] - test["DAYS_EMPLOYED"]
```

```python
train['고용비율'] = train['DAYS_EMPLOYED'] / train['DAYS_BIRTH'] # 인생 살면서 일한 비율
train['인당 평균 부양비'] = train['income_total'] / train['family_size'] 

test['고용비율'] = test['DAYS_EMPLOYED'] / test['DAYS_BIRTH'] # 인생 살면서 일한 비율
test['인당 평균 부양비'] = test['income_total'] / test['family_size'] 
```

```python
test['연봉'] = test['income_total'] / (test['근속연수'])
test['연봉'] = test['income_total'] / (test['근속연수'])
```

```python
train = train[(train['family_size'] <= 7)]
train = train.reset_index(drop=True)

test = test[(test['family_size'] <= 7)]
test = test.reset_index(drop=True)
```

# 인코딩

인코딩이 필요한 변수 범주형을 수치형으로
0.1 이렇게 있는 것을 원핫으로
0,1,2 이렇게 있는 것을 라벨인코딩으로
지금 인코딩을 해야하는 변수는 범주형 9개
인코딩을 잘 해야하는 변수는 house_type occyp_type
나머지는 인코딩을 해야 함


---


인코딩을 하는 방법에는 2가지가 있는데
1) 처음 방법은 머신러닝 라이브러리의 패키지를 써서 분리하는 방법 이다. 

sklearn 패키지안에 들어가 있어서 쓸 수 있지만, 

단점은 변환하고자 하는 컬럼명을 저렇게 다 써줘야 한다는 점이다. 

만약 하나가 오타가 난다거나 변환하고 싶은 범주형 변수가 많아지면 변수 이름들을 길게 나열해야 한다

2) get_dummies 사용 
한번에 처리할 수 있는 장점이 있지만 범주형 변수들을 따로 묶어서 변수를 만들어야 한다는 단점이 있다.


---
 house_type occyp_type 이 2개의 타입을 2개로 묶어서 원핫인코딩을 진행하고
 나머지 변수들이 그래서 인코딩이 필요하다는 것인가..?
 근데 house_type occyp_type 이 변수들의 타입이 6,7개라
 0으로 표현되므로 이러한 0들은 희소 벡터가 되며 벡터를 저장하기 위해 필요한 공간이 계속 늘어나므로 공간의 낭비가 발생하고 컴퓨터의 성능이 저하될 있음
그렇다면 라벨 인코딩을 진행하는 것이 더 좋은 방법인가
그럼 일단 라벨 인코딩으로  진행해보자!
