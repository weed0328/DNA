# 문제2. 수치형 - 범주형 전환문제.

데이터는 HHMM형태로 주어진 시간 데이터를 의미하고 있다.<br>
오전 11시의 경우 1100, 오후 4시의 경우 1600시인 24시 형태를 구성하고 있다.<br>

0601-1200까지의 시간을 오전으로 정의하고<br>
1201-2100까지의 시간을 오후라고 정의한다<br>
그 이외 시간대인 0000-0600, 2101-2400까지의 시간을 심야로 정의할때<br>

다음과 같이 input을 넣었을때, output을 반환하는 코드를 만들어라.<br>
[input]		[output]<br>
[1100]			[‘오전’] <br>
[1600, 2324]		[‘오후’, ‘심야’]<br>

출력은 함수의 형태로 진행되며 리스트의 값을 반환한다.<br>
input과 output의 길이는 가변 길이이며<br>
다음과 같은 형태의 예시를 지닌 함수를 만들어라<br>
ans = solution([1110, 2110, 0421, 1610])<br>
print(ans) # 오전, 심야, 심야, 오후 를 반환한다.

```python
def solution(list):
    a = []; b = []
    for i in list:
        if i[:2] == "00":
            i = i[2:]
            a.append(int(i))
        else:
            a.append(int(i))
    for i in a:
        if 601 <= i <= 1200:
            b.append("오전")
        if  1201 <= i <= 2100:
            b.append("오후")
        if  (0 <= i <= 600) or (2101 <= i <= 2400):
            b.append("심야")
    return b
ans = solution(["1110", "2110", "0421", "1610"])
print(ans)
```

3-3. data셋은 훌륭한 형태로 구성되어 있지만,<br>
cut(가공 수준)이 5개, color(색상)이 7개, clarity(흠집)가 8개의 범주를 가진 데이터셋이다.<br>
<br>
따라서 전처리가 필요한 단계로 보여지며,<br>
cut과 color 열에는 LabelEncoding을 진행하고<br>
clarity 열에는 OneHotEncoding을 진행하는것이 좋을것이라 판단하였다.<br>
<br>
사이킷런 혹은 판다스, 그외 기타 패키지나 사용자 정의 함수를 활용하여 라벨 인코딩을 진행하여라.<br>
이때 라벨 인코딩의 순서에는 변수의 순위를 고려한 인코딩을 진행하여라.<br>
<br>
cut은 가공수준의 단계를 나타내는 수준으로, 가장 좋은 단계부터 순서대로 기술하면 다음과 같다.<br>
Ideal<br>
Premium<br>
Very Good<br>
Good<br>
Fair<br>
<br>
여기서 clarity는 결점의 단계를 나타내는 범주인데, 가장 좋은 순서부터, 가장 나쁜 순서까지 다음과 같이 매핑된 결과이다.<br>
FL	Flawless<br>
IF	Internally Flawless<br>
VVS1 Very Very Slightly Included 1st<br>
VVS2 Very Very Slightly Included 2nd<br>
VS1 Very Slightly Included 1st<br>
VS2	Very Slightly Included 2nd<br>
SI1 Slightly Included 1st<br>
SI2	Slightly Included 2nd<br>
I1 - Included<br>


```python
# 변수의 순위를 고려해야 했기 때문에 부득이하게 다음과 같이 진행 함
cut = df2['cut']
encod_cut = []
for i in cut:
    if i == "Ideal":
        encod_cut.append(5)
    if i == "Premium":
        encod_cut.append(4)
    if i == "Very Good":
        encod_cut.append(3)
    if i == "Good":
        encod_cut.append(2)
    if i == "Fair":
        encod_cut.append(1)
    else:
        pass
encod_cut = pd.Series(encod_cut)
df2 = pd.concat([df2, encod_cut], axis=1)
df2.rename(columns={0:'encode_cut'}, inplace = True)
df2
```

```python
label = df2["color"]
label.unique()
le2 = LabelEncoder()
le2.fit(label)
label_encoded = le2.transform(label)
ledf = pd.DataFrame(label_encoded, columns=['label_encoded_color'])
df2 = pd.concat([df2, ledf], axis = 1)
```
